## Demo for microscope characterization: field dependent aberration

#### Setup environment

```python
import sys
sys.path.append("..")
from psflearning.psflearninglib import psflearninglib
from psflearning import io
from psflearning.makeplots import *
import tensorflow as tf
main_data_dir = io.param.load('datapath.yaml').main_data_dir
gpus = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
```

#### Load PSF model file from bead data

```python
resfile = main_data_dir + '/results/1ch_FD_bead/psfmodel_zernike_vector_FD_single.h5'
f,p = io.h5.load(resfile) # load result file
L = psflearninglib()
L.param = p
```

#### Display system characterizations


```python
showzernikemap(f,p)
```

```python
strehlratio = L.calstrehlratio(f)
```

```python
showpsfvsdata(f,p,index=50)
```

```python
fwhmx,fwhmy,fwhmz = L.calfwhm(f)
```
